```python
import pandas
pfm = pandas.read_csv("pfm_train.csv")
#pfm.describe()
pfm.sample(5)
```

```python
import seaborn
%matplotlib inline
seaborn.regplot(x='Age',y='Attrition', data=pfm)
```

```python
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(pfm['Attrition'])
lm = linear_model.LogisticRegression(max_iter=10000)
features = ['Age',
            'DistanceFromHome',
            'Education',
            'EmployeeNumber',
            'EnvironmentSatisfaction',
            'JobInvolvement',
            'JobLevel',
            'JobSatisfaction',
            'MonthlyIncome',
            'NumCompaniesWorked',
            'PercentSalaryHike',
            'PerformanceRating',
            'RelationshipSatisfaction',
            'StandardHours',
            'StockOptionLevel',
            'TotalWorkingYears',
            'TrainingTimesLastYear',
            'WorkLifeBalance',
            'YearsAtCompany',
            'YearsInCurrentRole',
            'YearsSinceLastPromotion',
            'YearsWithCurrManager']
x = pfm[features]
y = le.transform(pfm['Attrition'])
#y = pfm['Attrition']
#print(y)
scores = cross_val_score(lm,x,y,cv=5,scoring='accuracy')
print(np.mean(scores))
model = lm.fit(x,y)
#print(model.predict(34))
#print(model.intercept_, model.coef_)
```

```python
reader = pandas.read_csv("pfm_test.csv")
lenth = len(reader)
p0=0
p1=0
for i in range(lenth):
    tmp = [[reader.loc[i]['Age'],
            reader.loc[i]['DistanceFromHome'],
            reader.loc[i]['Education'],
            reader.loc[i]['EmployeeNumber'],
            reader.loc[i]['EnvironmentSatisfaction'],
            reader.loc[i]['JobInvolvement'],
            reader.loc[i]['JobLevel'],
            reader.loc[i]['JobSatisfaction'],
            reader.loc[i]['MonthlyIncome'],
            reader.loc[i]['NumCompaniesWorked'],
            reader.loc[i]['PercentSalaryHike'],
            reader.loc[i]['PerformanceRating'],
            reader.loc[i]['RelationshipSatisfaction'],
            reader.loc[i]['StandardHours'],
            reader.loc[i]['StockOptionLevel'],
            reader.loc[i]['TotalWorkingYears'],
            reader.loc[i]['TrainingTimesLastYear'],
            reader.loc[i]['WorkLifeBalance'],
            reader.loc[i]['YearsAtCompany'],
            reader.loc[i]['YearsInCurrentRole'],
            reader.loc[i]['YearsSinceLastPromotion'],
            reader.loc[i]['YearsWithCurrManager']]]
    #print(tmp)
    #print(model.predict(tmp))
    result = model.predict(tmp)
    if 0 == result:
        p0+=1
    else:
        p1+=1
    reader.loc[i,'Attrition'] = result
    #print(reader.loc[i]['BusinessTravel'])
print(p0)
print(p1)
reader.to_csv("pfm_test.csv")

```
