```python
import os
import sys
import json
import pandas  as pd
import numpy as np
```

```python
data = pd.read_table(filepath_or_buffer="./pfm_train.csv", sep=",")
print("是否有缺失值：",data.isnull().any().any())
```

# 去除肉眼可见无关因素

```python
data.drop(["EmployeeNumber"],axis=1,inplace=True)
```

# 去除固定值列

```python
drop_l =  []
for col in data.columns:
    length = len(data[col].unique())
    if length == 1:
        drop_l.append(col)
        
data.drop(drop_l,axis=1,inplace=True)
```

# 查看多重共线性问题，如果存在则删除其中一列

```python
corr = data.corr()
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(10,10))
sns.heatmap(corr,cmap="RdBu")
data.drop(["JobLevel"],axis=1,inplace=True)
```

# 构建新特征

```python
data.head()
```

```python
def build_feature(data):
    data["AgeEducation"] = data["Education"]*100/data["Age"]
    data["AgeIncome"] = data["MonthlyIncome"]*data["JobInvolvement"]*data["TotalWorkingYears"]//data["Age"]
    data["Satisfaction"] = data['JobSatisfaction']+data['EnvironmentSatisfaction']+data['RelationshipSatisfaction']
    return data
    
data = build_feature(data)
data = pd.get_dummies(data)
data.drop(["Education","MonthlyIncome","JobInvolvement","TotalWorkingYears","JobSatisfaction","EnvironmentSatisfaction","RelationshipSatisfaction"],axis=1,inplace=True)
data.head()
```

```python
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
y = data.Attrition
X = data.drop(["Attrition"], axis=1)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3,random_state = 149)
pd.Series(data["Attrition"]).value_counts()
params={
    'penalty':['l1', 'l2'],
    'C':np.arange(1,4.1,0.2),
}
estimator = linear_model.LogisticRegression(solver='liblinear')
grid = GridSearchCV(estimator,param_grid=params,cv=10)
grid.fit(X_train,y_train)
print('最好的参数',grid.best_params_)
score = grid.score(X_test,y_test)
print('在测试集的得分',score)
```

```python
model = linear_model.LogisticRegression(C=1.2, max_iter=1000)
a = model.fit(X_train, y_train)
pred = model.predict(X=X_test)
# print(pandas.Series(pred).value_counts())
score = metrics.accuracy_score(y_test, pred)
score_true = metrics.recall_score(y_test, pred)
score_false = metrics.recall_score(y_test, pred, pos_label=0)
print("理论准确率：", model.score(X_train, y_train))
print("准确率:", score)
print("正例覆盖率：", score_true)
print("负例覆盖率：", score_false)
```
